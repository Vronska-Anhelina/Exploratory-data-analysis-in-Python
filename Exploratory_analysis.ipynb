```python
print(df_countries.isna().sum())
print(df_countries.isna().sum()/df_countries.shape[0]*100) #There is one gap in each of the three columns, so they can be removed. Their share is only 0.4% out of 100%
df_country = df_countries.dropna()
print(df_country.isna().sum())
```

```python
print(df_events.isna().sum())
print(df_events.isna().sum()/df_events.shape[0]*100) # You can see that one column has 6% missing values ​​and the other has 0.1%. We can replace the country code rows with - Unklown (unknown). And y Units Sold is the median.
df_events["Units Sold"] = df_events["Units Sold"].fillna(df_events["Units Sold"].median())
df_events["Country Code"] = df_events["Country Code"].fillna("Unknown")
print(df_events.isna().sum())
```

```python
df = df_events.merge(df_products, left_on = "Product ID", right_on = "id", how = "left")
df = df.merge(df_country, left_on = "Country Code", right_on = "alpha-3", how = "left")
df.columns
columns_to_drop = [
    "id",
    "Country Code",
]
df = df.drop(columns=columns_to_drop, errors="ignore")
print(df)
```

```python
df["revenue"] = df["Units Sold"] * df["Unit Price"] #new revenue column

total_order_count = df["Order ID"].nunique()
total_revenue = df["revenue"].sum()
total_country = df["alpha-3"].nunique()
total_units_sold = df["Units Sold"].sum()
average_order_value = total_revenue / total_order_count
df["profit"] = (df["Unit Price"] - df["Unit Cost"]) * df["Units Sold"]
total_profit = df["profit"].sum()

print(f"Total number of orders: {total_order_count}")
print(f"Total revenue (excluding cost): {total_revenue}")
print(f"Number of unique countries: {total_country}")
print(f"Total number of units sold: {total_units_sold}")
print(f"Average check: {average_order_value}")
print(f"Total profit: {total_profit}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
category_profit = df.groupby("item_type")["profit"].sum().sort_values(ascending=False)
bar=category_profit.plot(kind="bar", figsize=(12, 6), color="slateblue")
plt.xlabel("Product Category",fontsize=14,
    color="black",)
plt.ylabel("Profit",
fontsize=14,
    color="black",)
plt.title("TOTAL PROFIT FOR CATEGORY",
fontsize=20,
    fontweight="bold",
    color="black",
    ha="center",)
plt.xticks(rotation=45)
for p in bar.patches:
    bar.annotate(f'{p.get_height():,.0f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points',
                fontsize=10)
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
df["total_cost"]=df["Unit Cost"]*df["Units Sold"]
analysis=df.groupby("item_type")[["revenue", "total_cost"]].sum().reset_index().sort_values(by="revenue",ascending=False)
df_melted=analysis.melt(id_vars="item_type",
                          value_vars=["revenue", "total_cost"],
                          var_name="Metric",
                          value_name="Amount")
plt.figure(figsize=(12, 6))
plot=sns.barplot(data=df_melted, x="item_type", y="Amount", hue="Metric", palette=["plum", "slateblue"])
handles,labels=plot.get_legend_handles_labels()
plot.legend(handles, ["Revenue", "Cost"], title="Indicators")

plt.title("TOTAL COST AND TOTAL REVENUE PER CATEGORY", fontsize=18, fontweight="bold")
plt.xticks(rotation=45)
plt.ylabel("Value")
plt.xlabel("Category")
for p in plot.patches:
    plot.annotate(f'{p.get_height():,.0f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points',
                fontsize=10)

plt.tight_layout()
plt.show()
```
