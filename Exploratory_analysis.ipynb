У датасеті event є такі колонки:

* Sales Channel - канал залучення. Судячи з перших 5-ти рядків, є Online i Offline магазини. Ця колонка розпізнається правильно (object).

* Unit Cost - Вартість одиниці.Собівартість. Тип розпізнається вірно (float64).

Датасет містить 1330 рядки та 10 стовпців, які я детально описала.

```python
print(df_countries.isna().sum())
print(df_countries.isna().sum()/df_countries.shape[0]*100) #Тут по одному пропуску в трьох стовпцях, тому їх можна видалити. Їхня доля лише 0.4% зі 100%
df_country = df_countries.dropna()
print(df_country.isna().sum())
```

```python
print(df_products.isna().sum())
print(df_products.isna().sum()/df_products.shape[0]*100) #Тут немає пропущених значень
```

```python
print(df_events.isna().sum())
print(df_events.isna().sum()/df_events.shape[0]*100) # Можна побачити, що в одному стовпці 6%  пропущених значень, а у другому 0.1%. Можемо замінити рядки в коді країни на - Unklown (невідомо). А  y Units Sold медіаною.
df_events["Units Sold"] = df_events["Units Sold"].fillna(df_events["Units Sold"].median())
df_events["Country Code"] = df_events["Country Code"].fillna("Unknown")
print(df_events.isna().sum())
```

```python
df_events["Order Date"] = pd.to_datetime(df_events["Order Date"])
df_events["Ship Date"] = pd.to_datetime(df_events["Ship Date"])
df_events.dtypes
```

```python
df = df_events.merge(df_products, left_on = "Product ID", right_on = "id", how = "left")
df = df.merge(df_country, left_on = "Country Code", right_on = "alpha-3", how = "left")
df.columns
columns_to_drop = [
    "id",
    "Country Code",
]
df = df.drop(columns=columns_to_drop, errors="ignore")
print(df)
```

```python
df["revenue"] = df["Units Sold"] * df["Unit Price"] #нова колонка з виручкою

total_order_count = df["Order ID"].nunique()
total_revenue = df["revenue"].sum()
total_country = df["alpha-3"].nunique()
total_units_sold = df["Units Sold"].sum()
average_order_value = total_revenue / total_order_count
df["profit"] = (df["Unit Price"] - df["Unit Cost"]) * df["Units Sold"] #Прибуток
total_profit = df["profit"].sum()

print(f"Загальна кількість замовлень: {total_order_count}")
print(f"Загальна виручка(без урахування собівартості): {total_revenue}")
print(f"Кількість унікальних країн: {total_country}")
print(f"Загальна кількість проданих одиниць: {total_units_sold}")
print(f"Середній чек: {average_order_value}")
print(f"Загальний прибуток: {total_profit}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
category_profit = df.groupby("item_type")["profit"].sum().sort_values(ascending=False)
bar=category_profit.plot(kind="bar", figsize=(12, 6), color="slateblue")
plt.xlabel("Product Category",fontsize=14,
    color="black",)
plt.ylabel("Profit",
fontsize=14,
    color="black",)
plt.title("TOTAL PROFIT FOR CATEGORY",
fontsize=20,
    fontweight="bold",
    color="black",
    ha="center",)
plt.xticks(rotation=45)
for p in bar.patches:
    bar.annotate(f'{p.get_height():,.0f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points',
                fontsize=10)
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
df["total_cost"]=df["Unit Cost"]*df["Units Sold"]
analysis=df.groupby("item_type")[["revenue", "total_cost"]].sum().reset_index().sort_values(by="revenue",ascending=False)
df_melted=analysis.melt(id_vars="item_type",
                          value_vars=["revenue", "total_cost"],
                          var_name="Metric",
                          value_name="Amount")
plt.figure(figsize=(12, 6))
plot=sns.barplot(data=df_melted, x="item_type", y="Amount", hue="Metric", palette=["plum", "slateblue"])
handles,labels=plot.get_legend_handles_labels()
plot.legend(handles, ["Дохід (Revenue)", "Витрати (Cost)"], title="Показники")

plt.title("TOTAL COST AND TOTAL REVENUE PER CATEGORY", fontsize=18, fontweight="bold")
plt.xticks(rotation=45)
plt.ylabel("Value")
plt.xlabel("Category")
for p in plot.patches:
    plot.annotate(f'{p.get_height():,.0f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points',
                fontsize=10)

plt.tight_layout()
plt.show()
```
